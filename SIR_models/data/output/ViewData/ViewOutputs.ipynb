```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
fig = plt.figure(figsize=(16,8))

data = np.loadtxt("casesWithPop.csv", delimiter=",")
day=data[1:,0] #first column is day
regions=["Day","Ontario","Algoma","Chatham-Kent","Hamilton","Brant","Durham","Eastern Ontario","Grey Bruce","Haldimand-Norfolk","Haliburton Kawartha Pine Ridge","Halton","Hastings Prince Edward","Huron Perth","Kingston Frontenac Lennox Addington","Lambton","Leeds Grenville Lanark","Middlesex-London","Niagara","North Bay Parry Sound","Northwestern","Ottawa","Peel","Peterborough","Porcupine","Sudbury","Waterloo","Renfrew","Simcoe Muskoka","Southwestern","Thunder Bay","Timiskaming","Toronto","Wellington-Dufferin-Guelph","Windsor-Essex","York"]

popn=data[0,:] #first row is population

plt.grid()
for iregion in range (1,len(regions)): # starting at column 1 since column 0 is days   
    if 200e3 < popn[iregion] < 215e3 :
        newInfected = data[1:,iregion] #column

        plt.plot(day,newInfected,label=regions[iregion])
        
plt.title("Infected by region ON")
plt.xlabel("Days (day 1 is Jan 15)")
#plt.yscale("log")
plt.ylim(0,16)
plt.xlim(40,140)
plt.legend()
plt.show()
```

```python
fig = plt.figure(figsize=(16,8))

data = np.loadtxt("UKregions.csv", delimiter=",")
day=data[:,0] #first column is day
regions=["Day","East Midlands 5M","East of England 6M","London 9M","North East 2.5M","North West 7M","South East 9M","South West 5.5M","West Midlands 6M","Yorkshire and The Humber 5.5M"]

plt.grid()
for iregion in range (1,len(regions)):    
    newInfected = data[:,iregion] #column

    plt.plot(day,newInfected,label=regions[iregion])

plt.title("Infected by region UK (population ranges 2M-8M)")
plt.xlabel("Days (day 1 is Jan 30)")
#plt.yscale("log")
plt.legend()
plt.show()
```
