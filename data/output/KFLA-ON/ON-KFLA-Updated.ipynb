```python
import numpy as np
import matplotlib.pyplot as plt
```

```python

dataK = np.loadtxt("kingstonData.csv", delimiter=",")
dayK=dataK[:,0]
newInfectedK = dataK[:,1] #column

dataO = np.loadtxt("allOntario.csv", delimiter=",")
dayO=dataO[:,0]
newInfectedO = dataO[:,1] #column

# Kingston
fig = plt.figure(figsize=(10,8))

plt.subplot(2,1,1)
plt.grid()
## real data ##
plt.plot(dayK,newInfectedK,label="KFL&A data")
## ######### ##
plt.title("Newly infected in KFL&A region")
plt.xlabel("Time in days")
plt.ylabel("Number of people")
plt.show()

# Ontario
fig = plt.figure(figsize=(12,12))

plt.subplot(2,1,1)
plt.grid()
## real data ##
plt.plot(dayO,newInfectedO,label="Ontario data")
## ######### ##
plt.title("Newly infected in Ontario")
plt.xlabel("Time in days")
plt.ylabel("Number of people")
plt.show()
```
